# Idealized Case 2: Two crossing Blobs

This tutorial shows the most important steps of tracking with `tobac` using an idealized case:

1. [Data Generation](#1.-Data-Generation)

## Data Generation

```python
import tobac
import numpy as np
import matplotlib.pyplot as plt
import datetime
import xarray as xr
import seaborn as sns

sns.set_context("talk")
```

```python
x_min, y_min, x_max, y_max, = (
    0,
    0,
    10,
    10,
)
t_min, t_max = 0, 10

x = np.linspace(x_min, x_max, 100)
y = np.linspace(y_min, y_max, 100)
t = np.linspace(t_min, t_max, 20)

tt, yy, xx = np.meshgrid(t, y, x, indexing="ij")
```

```python
def create_wandering_blob(
    x_0, y_0, v_x, v_y, sigma=0.1, t_create=t_min, t_vanish=t_max
):

    exponent = (xx - x_0 - v_x * (tt - t_create)) ** 2 + (
        yy - y_0 - v_y * (tt - t_create)
    ) ** 2
    feature_alive = np.logical_and(tt >= t_create, tt <= t_vanish)
    blob = np.where(feature_alive, np.exp(-exponent / sigma), 0)

    return blob


def create_xarray(array):

    t_0 = np.datetime64("2022-04-01T00:00")
    t_delta = np.timedelta64(1, "m")
    time = np.array([t_0 + i * t_delta for i in range(len(array))])

    coords = {"time": time, "projection_x_coordinate": x, "projection_y_coordinate": y}

    attributes = {"units": ("m s-1")}

    data = xr.DataArray(data=array, coords=coords, attrs=attributes)
    return data
```

```python
blob_1 = create_wandering_blob(0, 0, 1, 1)
blob_2 = create_wandering_blob(10, 0, -1, 1)
blob_mask = blob_1 > blob_2
blob = np.where(blob_mask, blob_1, blob_2)

data = create_xarray(blob)
```

```python
%matplotlib inline
data.plot(
    cmap="viridis",
    col="time",
    col_wrap=5,
    x="projection_x_coordinate",
    y="projection_y_coordinate",
    size=5,
)
```

## Influence of Tracking Method

```python
%%capture
dxy, dt = tobac.utils.get_spacings(data, grid_spacing=x_max / 200)
features = tobac.themes.tobac_v1.feature_detection_multithreshold(
    data, dxy, threshold=0.9
)
```

```python
%matplotlib inline
d_max = 0.4

track_1 = tobac.themes.tobac_v1.linking_trackpy(
    features, data, dt, dxy, d_max=d_max, method_linking="random", subnetwork_size=2
)

track_2 = tobac.themes.tobac_v1.linking_trackpy(
    features, data, dt, dxy, d_max=d_max, method_linking="predict", subnetwork_size=2
)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(20, 10))

for i, cell_track in track_1.groupby("cell"):
    cell_track.plot.scatter(
        x="projection_x_coordinate",
        y="projection_y_coordinate",
        ax=ax1,
        marker="o",
        label="cell {0}".format(int(i)),
    )
ax1.legend()
ax1.set_title("random")

for i, cell_track in track_2.groupby("cell"):

    cell_track.plot.scatter(
        x="projection_x_coordinate",
        y="projection_y_coordinate",
        ax=ax2,
        marker="o",
        label="cell {0}".format(int(i)),
    )
ax2.legend()
ax2.set_title("predict")
```

### Search Radius 

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(22, 10), sharex=True, sharey=True)
radius = int(d_max / dxy)

for frame in [8, 9, 10]:

    features.where(features["frame"] == frame).plot.scatter(
        x="hdim_1", y="hdim_2", color="blue", ax=ax1
    )

    features.where(features["frame"] == frame).plot.scatter(
        x="hdim_1", y="hdim_2", color="blue", ax=ax2
    )

position_0 = (
    features.where(features["frame"] == 8).dropna(dim="index")["hdim_1"][0],
    features.where(features["frame"] == 8).dropna(dim="index")["hdim_2"][0],
)

position_1 = (
    features.where(features["frame"] == 9).dropna(dim="index")["hdim_1"][0],
    features.where(features["frame"] == 9).dropna(dim="index")["hdim_2"][0],
)
position_2 = (
    features.where(features["frame"] == 10).dropna(dim="index")["hdim_1"][1],
    features.where(features["frame"] == 10).dropna(dim="index")["hdim_2"][1],
)

ax1.plot([position_1[0], position_1[0] - radius], [position_1[1], position_1[1]], "k--")
ax1.plot(
    position_1[0],
    position_1[1],
    marker="o",
    color="red",
    markersize=10,
    linestyle="",
    label="feature position",
)
circle_1 = plt.Circle(position_1, radius, fill=False, label="Search Radius")
ax1.add_patch(circle_1)
ax1.set_title("Random")
ax1.legend()

ax2.plot([position_2[0], position_2[0] - radius], [position_2[1], position_2[1]], "k--")
ax2.plot(
    position_1[0],
    position_1[1],
    marker="o",
    color="red",
    markersize=10,
    linestyle="",
    label="feature position",
)
ax2.plot(
    position_2[0],
    position_2[1],
    marker="x",
    color="black",
    markersize=15,
    linestyle="",
    label="predicted next position",
)
circle_2 = plt.Circle(position_2, radius, fill=False, label="Search Radius")
ax2.add_patch(circle_2)
ax2.arrow(
    position_0[0],
    position_0[1],
    position_1[0] - position_0[0],
    position_1[1] - position_0[1],
    head_width=0.5,
    length_includes_head=True,
    color="black",
)
ax2.arrow(
    position_1[0],
    position_1[1],
    position_2[0] - position_1[0],
    position_2[1] - position_1[1],
    head_width=0.5,
    length_includes_head=True,
    color="black",
    linestyle="dotted",
)
ax2.set_title("Predict")
ax2.legend()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

position_0 = (
    features.where(features["frame"] == 9).dropna(dim="index")["hdim_1"][0],
    features.where(features["frame"] == 9).dropna(dim="index")["hdim_2"][0],
)
plt.text(position_0[0] - 0.25, position_0[1] - 0.15, "Position 0")

position_1 = (
    features.where(features["frame"] == 9).dropna(dim="index")["hdim_1"][1],
    features.where(features["frame"] == 9).dropna(dim="index")["hdim_2"][1],
)
plt.text(position_1[0] - 0.25, position_1[1] + 0.15, "Position 1")
position_2 = (
    features.where(features["frame"] == 10).dropna(dim="index")["hdim_1"][0],
    features.where(features["frame"] == 10).dropna(dim="index")["hdim_2"][0],
)
plt.text(position_2[0] - 0.25, position_2[1] - 0.15, "Position 2")
position_3 = (
    features.where(features["frame"] == 10).dropna(dim="index")["hdim_1"][1],
    features.where(features["frame"] == 10).dropna(dim="index")["hdim_2"][1],
)
plt.text(position_3[0] - 0.25, position_3[1] + 0.15, "Position 3")

ax.plot(
    [position_0[0], position_2[0]],
    [position_0[1], position_2[1]],
    color="blue",
    label="Option 1",
)
ax.plot([position_1[0], position_3[0]], [position_1[1], position_3[1]], color="blue")

ax.plot(
    [position_0[0], position_3[0]],
    [position_0[1], position_3[1]],
    color="red",
    label="Option 2",
)
ax.plot([position_1[0], position_2[0]], [position_1[1], position_2[1]], color="red")

for frame in [9, 10]:

    features.where(features["frame"] == frame).plot.scatter(
        x="hdim_1", y="hdim_2", color="black", ax=ax
    )

plt.title("Options for Linking")
plt.legend()
```

### Sum of Distances for Linking

```python
# random, option 1

np.sqrt(np.sum((np.array(position_0) - np.array(position_2)) ** 2)) + np.sqrt(
    np.sum((np.array(position_1) - np.array(position_3)) ** 2)
)
```

```python
# random, option 2

np.sqrt(np.sum((np.array(position_0) - np.array(position_3)) ** 2)) + np.sqrt(
    np.sum((np.array(position_1) - np.array(position_2)) ** 2)
)
```

```python
# predict, option 1

np.sqrt(np.sum((np.array(position_3) - np.array(position_2)) ** 2)) + np.sqrt(
    np.sum((np.array(position_2) - np.array(position_3)) ** 2)
)
```

```python
# predict, option 2

np.sqrt(np.sum((np.array(position_3) - np.array(position_3)) ** 2)) + np.sqrt(
    np.sum((np.array(position_2) - np.array(position_2)) ** 2)
)
```
